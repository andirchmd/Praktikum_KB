## Memvisualisasikan sebuah Dataset Sentiment


---

Dataset: [PRDECT-ID: Indonesian Emotion Classification | Kaggle](https://www.kaggle.com/datasets/jocelyndumlao/prdect-id-indonesian-emotion-classification/data)

Untuk menganalisis dan memvisualisasikan dataset ini saya menggunakan library `pandas`, `matplotlib (pyplot)`, dan `seaborn`

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Baca dan simpan datasetnya dengan nama `df` dan data siap digunakan

```python
df = pd.read_csv("PRDECT-ID Dataset.csv")
df.head()
```

### Ketentuan Visualisasi Data


* Gunakan minimal 4 tipe visualiasi data.
* Wajib menggunakan pie chart dan heatmap.
* Kesimpulan dari visualisasi data dituliskan menggunakan cell  Markdown.
* Gunakan title, xlabel, dan ylabel pada setiap visualisasi.
* Semakin informatif visualisasi, semakin tinggi nilai posttest.



**Visualisasi 1 Pie Chart**

---
Visualisasi Pie Chart ini menampilkan proporsi emosi pada dataset.

> Dataset ini memiliki variasi emosi, dengan "Happy" sebagai emosi yang paling dominan diikuti oleh "Sadness", "Fear" lalu "Anger" memiliki distribusi paling rendah. Meskipun emosi Happy terlihat dominan, pada distribusi sentimen kita dapat melihat bahwa 52% customer memberi sentimen negatif.

```python
emotion_counts = df['Emotion'].value_counts()
sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribusi Emosi')
plt.show()

plt.figure(figsize=(4, 4))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribusi Sentimen')
plt.axis('equal')
plt.show()
```

**Visualisasi 2 Heatmap**

---
Visualisasi Heatmap ini menampilkan korelasi pada dataset.

```python
plt.figure(figsize=(8, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi')
plt.show()
```

**Visualisasi 3 Histogram**

---
Visualisasi Histogram ini menampilkan Jumlah Review per Emosi

> Melalui visualisasi ini kita dapat melihat bahwa lebih dari 1750 customer yang memberi review memiliki sentiment *Happy* yang diikuti oleh *Sadness*, *Fear*, *Love* dan yang paling sedikit yaitu *Anger*

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Emotion'], bins=len(df['Emotion'].unique()), kde=False)
plt.title('Jumlah Review per Emosi')
plt.xlabel('Emosi')
plt.ylabel('Jumlah Review')
plt.show()
```

**Visualisasi 4 Box Plot**

---
Visualisasi Box Plot ini menampilkan jumlah panjang teks untuk setiap emosi pada dataset.

> Dengan informasi yang kita dapat, distribusi panjang teks didapat dari emosi "Fear" sangat panjang dibanding rata-rata panjang teks pada emosi lainnya sedangkan emosi "Happy" memiliki panjang teks yang paling sedikit.

```python
plt.figure(figsize=(10, 6))
df['Text_Length'] = df['Customer Review'].apply(lambda x: len(str(x)))
sns.boxplot(y='Emotion', x='Text_Length', data=df)
plt.title('Distribusi Panjang Teks per Emosi')
plt.ylabel('Emosi')
plt.xlabel('Panjang Teks')
plt.show()
```

**Visualisasi 5 Bar Chart**

---
Visualisasi Bar Chart ini menampilkan Rata-rata Customer Rating berdasarkan Emosi
> Pada Visualisasi bar chart ini kita dapat melihat kalau customer dengan emosi Happy dan Love lebih sering memberi rating tinggi dibanding dengan emosi lainnya.

```python
average_rating = df.groupby('Emotion')['Customer Rating'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=average_rating.index, y=average_rating.values, palette="viridis")
plt.title("Rata-Rata Customer Rating Berdasarkan Emosi")
plt.xlabel("Emosi")
plt.ylabel("Rata-Rata Customer Rating")
plt.show()
```
