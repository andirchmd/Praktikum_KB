**Analisis 3**

---
* Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
* Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya.
* Tampilkan isi dari dictionary tersebut sebagus mungkin

```python
angka = df.select_dtypes(include=['number'])

# buat dictionarynya
info_dict = {
    'Jumlah Record': len(df),
    'Jumlah Attribute': len(df.columns),
    'Nama Attribute': df.columns.tolist(),
    'Jumlah Attribute Angka': len(angka.columns)
}

# tampilkan dictionary
print("="*45)
print(" "*14+"Informasi Dataset")
print("="*45)
for key, value in info_dict.items():
    if isinstance(value, list): ## jika isi value == list
        value ='\n                         '.join(map(str, value)) ## mengubah list menjadi string dan pisahkan dengan \n dan spasi
        print(f"{key:<23}: {value}")
    else:
        print(f"{key:<23}: {value}")
print("=" * 45)
```

**Analisis 5**

---
Menampilkan jumlah nilai dari semua attribute yang bernilai object

> Fungsi `.nunique()` berguna untuk menghitung jumlah nilai unik pada objek yang memiliki nilai berbeda dan mengabaikan nilai duplikat.

reference: [nunique docs ](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nunique.html)

```python
objek = df.select_dtypes(include=['object'])
jumlah = objek.nunique()

table = pd.DataFrame({
    'Nama Atribut': jumlah.index,
    'Total Nilai': jumlah.values})

display(table)
```
